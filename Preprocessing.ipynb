```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
files = ["learn_dataset.csv", "learn_dataset_emp_contract.csv", "learn_dataset_job.csv", "learn_dataset_sport.csv"]

data = [] 

for file in files:
    data.append(pd.read_csv(file, index_col="UID"))
```

```python
df = pd.concat(data, axis=1, ignore_index=False)
```

```python
#df = df.drop(["OCCUPATION_42", "JOB_DEP","household_type", "Insee", "emp_contract",  "eco_sect", "employee_count", "Work_description", "Sports" ], axis=1)
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.model_selection import GridSearchCV
```

```python
X_train = df.drop(["target"], axis=1)
y_train = df["target"]
```

```python
X_train
```

```python
numerical_features   = make_column_selector(dtype_include=np.number)
categorical_features = make_column_selector(dtype_exclude=np.number) 
```

```python
numerical_pipeline   = make_pipeline(SimpleImputer(strategy="mean"), 
                                     StandardScaler())
categorical_pipeline = make_pipeline(SimpleImputer(strategy="most_frequent"), 
                                     OneHotEncoder())
```

```python
prepocessor = make_column_transformer((numerical_pipeline, numerical_features),
                                     (categorical_pipeline, categorical_features))
```

```python
param_grid = {
    "learning_rate" : ["adaptive", "invscaling"],
    "loss":["squared_error", "huber"]
}
model = make_pipeline(prepocessor, GridSearchCV(SGDRegressor(),
                                                param_grid,
                                               cv=3,
                                               refit=True
                                               ))
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```
